```python
import numpy as np
import cv2
import skimage as ski
import os
from PIL import Image
```

```python
#Convertir imagenes del dataset a su version Fast Fourier Transform (FFT) invertido
```

```python
def out_FFT(img):
    img = ski.color.rgb2gray(img)
    img = ski.img_as_ubyte(img)

    dft = np.fft.fft2(img, axes=(0, 1))
    dft_shift = np.fft.fftshift(dft)
    mag = np.abs(dft_shift)
    spec = np.log(mag) / 20
    return spec
```

```python
# pasar imagenes del folder train/IA a FFT
path = r"train\IA"
out_path = r"fft_train\IA"

for i, image_path in enumerate(os.listdir(path)):
    # leemos cada imagen con skimage
    input_path = os.path.join(path, image_path)
    image = Image.open(input_path).convert("RGB")

    # aplicamos FFT invertido
    fft = out_FFT(image)

    # normalizamos la imagen porque si no skimage no la quiere guardar como jpg
    image_norm = ski.exposure.rescale_intensity(fft, out_range=(0, 1))
    image_norm = (image_norm * 255).astype(np.uint8)
    
    # se re-escala a 256x256
    image = ski.transform.resize(image_norm, (256, 256), anti_aliasing=False, order=0)


    # guardamos imagenes en nuevo folder
    fullpath = os.path.join(out_path, str(i) + image_path)
    
    ski.io.imsave(fullpath, image) 
```
